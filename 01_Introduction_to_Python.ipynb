### Variables and Data Types

Python supports various data types
- Text type: str
- Numeric types: int, float, complex
- Boolean type:	bool
- Sequence types:	list, tuple
- Mapping type:	dict
- Set types:	set, frozenset
- Binary types:	bytes, bytearray, memoryview
- None type:	NoneType

#### Text, numeric and boolean data types

```python
name = "Fredl" # string
age = 42 # integer
height = 1.8 # float
is_student = True # boolean

print(name + " is " + str(age) + " years old, " + str(height) + " m tall and is a student: " + str(is_student))
```

There are further ways to produce the same output e.g. with a f-string

```python
print(f"{name} is {age} years old, {height} m tall and is a student: {is_student}")
```

or with the format method

```python
print("{} is {} years old, {} m tall and is a student: {}".format(name, age, height, is_student))
```

In case of big or very small numbers an underscore (_) can be use in numerical literals.
Furthermore the exponential notation can be used

```python
a = 1_000_000
b = 1e6 # stored as a float

c = 0.000_001
d = 1e-6 

print(a, b, c, d)
```

##### Complex numbers

Complex numbers can easily typed out of the box

```python
c = 10 - 5j # complex(10, -5)
```

Extract the real and imaginary part from the complex number

```python
r = c.real
i = c.imag

print(f"Real: {r}, Imaginary: {i}")
```

Complex conjugated number

```python
c_conjugated = c.conjugate()

print(f"Conjugated: {c_conjugated}")
```

Calculate the absolute value of the complex number

```python
c_absolute = abs(c)

print(f"Absolute: {c_absolute}")
```

To check whether or not Python fools us, the absolute value of a complex number is calculated by:
$$|c| = \sqrt{\Re(c)^2 + \Im(c)^2}$$

```python
import math

math.sqrt(c.real**2 + c.imag**2)
```

#### Sequence, mapping and set data types

##### List

Lists are used to store multiple items in a single variable.

```python
my_list = [1, 2, 3, 4, 5]
print(my_list)
```

List items are indexed, the first item has index [0], the second item has index [1] etc. The last item is at [-1], the second last at [-2].

```python
print(f"First element: {my_list[0]}, second element: {my_list[1]}, last element: {my_list[-1]}, second last element: {my_list[-2]}")
```

A list is changeable and allows duplicates

```python
my_list[2] = 4
print(my_list)
```

List items can be of any data type

```python
my_list =[1, "apple", 3.14, True, [1, 2, 3]]
print(my_list)
```

##### Tuple

Tuples are similiar to lists but tuples are **unchangable**

```python
my_tuple = (1, 2, 3, 4, 5)
my_tuple[2] = 4 # TypeError: 'tuple' object does not support item assignment
```

##### Dictionary

Dictionaries are used to store data values in key:value pairs. A dictionary is a collection which is ordered, changeable and do not allow duplicates

```python
my_dict = {
  "substance": "Ethanol",
  "formula": "C2H5OH",
  "molar_mass": 46.07,
}

print(my_dict)
```

Dictionaries are indexed by its keys

```python
molar_mass = my_dict["molar_mass"]
substance = my_dict["substance"]

print(f"Molar mass of {substance} is {molar_mass} g/mol")
```

Keys and values can be extraced into lists

```python
keys = my_dict.keys()
values = my_dict.values()

print(f"The keys are: {keys} and the values are: {values}")
```

##### Memory view

```python
# Create a bytearray object
data = bytearray(b"Hello, World!")

# Create a memoryview of the bytearray
mv = memoryview(data)

# Modify the original bytearray via memoryview
mv[7:12] = b'Bytes'

print(data)
```

```python
import time

for n in (100_000, 200_000, 300_000, 400_000):
    data = b'x'*n
    start = time.time()
    b = data
    while b:
        b = b[1:]
    print(f'For {n} bytes {time.time() - start:0.3f}')
```

```python
for n in (100_000, 200_000, 300_000, 400_000):
    data = b'x'*n
    start = time.time()
    b = memoryview(data)
    while b:
        b = b[1:]
    print(f'For {n} bytes {time.time() - start:0.3f}')
```
